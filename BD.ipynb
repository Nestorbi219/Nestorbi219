Таблица Readers:

```python
from faker import Faker

import random

fake = Faker()

for i in range(1000):
    name = fake.name()
    address = fake.address().replace('\n', '\\n')
    phone_number = ''.join([str(random.randint(0, 9)) for _ in range(10)])
    email = fake.email()
    birthdate = fake.date_of_birth().strftime('%Y-%m-%d')
    reader_id = i + 1

    print("INSERT INTO Readers (НомерЧитБилета, ФИО, Адрес, Телефон, ЭлПочта, ДатаРождения) VALUES (%s, '%s', '%s', '%s', '%s', '%s')" % (reader_id, name, address, phone_number, email, birthdate))
```

Таблица Rent:

```python
from faker import Faker
import random
from datetime import datetime, timedelta

fake = Faker()

for i in range(1000):
    book_id = random.randint(1, 1000)
    reader_id = random.randint(1, 1000)
    rent_date = fake.date_between(start_date='-1y', end_date='today').strftime('%Y-%m-%d')
    rent_cost = random.randint(100, 1000)
    return_date = (datetime.strptime(rent_date, '%Y-%m-%d') + timedelta(days=random.randint(30, 60))).strftime('%Y-%m-%d')
    rent_id = i + 1

    print("INSERT INTO Rent (АрендаID, КнигаID, ЧитательID, ДатаАренды, СтоимостьАренды, ДатаВозврата) VALUES (%s, %s, %s, '%s', %s, '%s')" % (rent_id, book_id, reader_id, rent_date, rent_cost, return_date))
```
